### Levy

$\large f(x,y) = \sin^2 3 \pi x + (x-1)^2 (1 + \sin^2 3 \pi x) + (y-1)^2 (1 + \sin^2 3 \pi y)$

### Michalewicz

$\large f(x,y) = \sin(x_1) \; \sin( 1 . x_1^2 / \pi ) ^ {2 . \; 0.5} + \sin(x_2) \; \sin( 2 x_2^2 / \pi ) ^ {2 . \; 0.5} $

```python
of.show_function(of.michalewicz2, [-5,5])
```
