# Simulated annealing

```python
import sys
sys.path.append('D:/source/repos')
from utilities.std_imports import *
from scipy.optimize import dual_annealing
```

### Define a function and interval (plot is just projection)

```python
d=3
func = lambda x: np.sum(x*x - d*np.cos(2*np.pi*x)) + d*np.size(x)
lwr = -5.12
upr = 5.12

X = np.linspace(start=lwr, stop=upr, num=400)
Y = [func(x) for x in X]
plt.figure(figsize=[16,8])
plt.plot(X,Y);
```

### Calculate and show results

```python
bounds = list(zip([lwr]*d, [upr]*d))
res = dual_annealing(func, bounds=bounds, seed=1234)
res.x, res.fun
```
